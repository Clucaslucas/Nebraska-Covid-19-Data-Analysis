```python
import datetime
with open("Nebraska Positive Cases Per Date-1.txt", "r") as f:
    data = f.readlines()
nums = [int(num.strip()) for num in data]
total = sum(nums)
average = round(total / len(nums), 2)
maxNum = max(nums)
indexMax = nums.index(maxNum)

start = datetime.date(2020, 3, 8)

if indexMax >= 1:
    indexMax += 1
if indexMax >= 3:
    indexMax += 1
print("Total cases: ", total,"Average: ", average,"Max: ", maxNum,"Index of Max:", indexMax)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
##Reading data
data = pd.read_csv("Nebraska_Coronavirsu_PositiveCases-1.csv")
print(data.head())

##Graph
first_day_cases = data.iloc[0, 1] 
print(f"Number of new cases reported on the first day of screening: {first_day_cases}")
plt.figure(figsize=(10, 5))
plt.plot(data['Date'], data['Total_Positive_Cases'])
plt.title('CoronaVirus Positive Cases in Nebraska Over time')
plt.xlabel('Date')
plt.ylabel('Total Number of Positive Cases')
plt.xticks(rotation=45) 
plt.tight_layout()  
plt.show()
```

```python
##a
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("NewYork_Nebraska_PositiveCases-1.csv")
print(data.head())
print(data.tail())
##b
nebraskaTotal = data['NEW_COVID_CASE_COUNT_Nebraska'].sum()
print(f"Total cases in Nebraska: {nebraskaTotal}")
##c
newYorkTotal = data['NEW_COVID_CASE_COUNT_NewYork'].sum()
print(f"Total cases in New York: {newYorkTotal}")
##d
plt.figure(figsize=(10, 5))
plt.plot(data['Date'], data['total_NE'], label="Nebraska", color="blue")
plt.title('Total COVID-19 Cases in Nebraska Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(data['Date'], data['total_NY'], label="New York", color="red")
plt.title('Total COVID-19 Cases in New York Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()
##e
avg_nebraska = data['NEW_COVID_CASE_COUNT_Nebraska'].mean()
avg_newyork = data['NEW_COVID_CASE_COUNT_NewYork'].mean()
print(f"Average daily cases in Nebraska: {avg_nebraska:.2f}")
print(f"Average daily cases in New York: {avg_newyork:.2f}\n")
##f
fiveNumNebraska = data['NEW_COVID_CASE_COUNT_Nebraska'].describe()[['min', '25%', '50%', '75%', 'max']]
fiveNumNewYork = data['NEW_COVID_CASE_COUNT_NewYork'].describe()[['min', '25%', '50%', '75%', 'max']]
print("Five-number summary for Nebraska:", fiveNumNebraska)
print("Five-number summary for New York:", fiveNumNewYork)
```
